### ðŸ“‹ Outlier Summary with IQR Method

We define a function to detect **outliers** using the **Interquartile Range (IQR)** method:

- For each non-binary numeric column, values beyond 1.5Ã—IQR from Q1 or Q3 are flagged as outliers.
- The output provides a summary including:
  - Number and percentage of outliers
  - Minimum and maximum outlier values

This helps prioritize which features may need **outlier handling** based on their impact.

```python
def detect_outliers_iqr(df, cols=None):
    if cols is None:
        cols = df.select_dtypes(include=np.number).columns
    
    outlier_summary = {}

    for col in cols:
        if df[col].nunique() <= 2:  # skip binary
            continue
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR

        outliers = df[(df[col] < lower) | (df[col] > upper)][col]
        outlier_summary[col] = {
            "count": outliers.count(),
            "percent": round(100 * outliers.count() / len(df), 2),
            "min": outliers.min() if not outliers.empty else None,
            "max": outliers.max() if not outliers.empty else None,
        }
    return pd.DataFrame(outlier_summary).T.sort_values("percent", ascending=False)

# Run on df_2
outlier_report = detect_outliers_iqr(df_2)
outlier_report
```

### Refined Outlier Detection

We apply a **stricter outlier rule** using **3Ã—IQR** to reduce false positives from heavy-tailed or sparse variables.

- For near-constant or zero-heavy features, we use the **99.5th percentile** as a cutoff.
- This approach flags only the most extreme values for further review or handling.

```python
outliers = {}
for col in non_binary_columns_names:
    s = df_2[col].dropna()
    q1, q3 = s.quantile([0.25, 0.75])
    iqr = q3 - q1

    if iqr == 0:
        # Discrete / zero-heavy: use extreme percentile cutoff, otherwise columns with many zeroes will present many outliers even if unique number of values very low
        upper = s.quantile(0.995)  # 99.5th percentile
        mask = s > upper           # only flag extreme high values
    else:
        lower = q1 - 3*iqr         # stricter than 1.5Ã—IQR
        upper = q3 + 3*iqr
        mask = (s < lower) | (s > upper)

    if mask.any():
        outliers[col] = df_2.loc[mask.index[mask], [col]]

# List of columns with outliers
cols_with_outliers = list(outliers.keys())

# Display results
for col, data in outliers.items():
    print(f"\n{col} â€” {len(data)} outliers")
    display(data)
```

```python
# This block identifies clients who have extreme values (outliers) in more than one column.
# Why: A client with multiple outlier fields may be an anomaly or data error that could distort analysis.
# Method: 
#   1. For each column in cols_with_outliers, apply Tukey's rule (3 Ã— IQR) to flag outlier values as True.
#   2. Count how many columns are flagged True for each client.
#   3. Collect Client_IDs with more than one True flag â€” these may be candidates for removal or deeper review.
# Output: 
#   - multi_outlier_clients: list of Client_IDs meeting the >1 outlier criterion.
#   - multi_outlier_df: DataFrame showing outlier flags for each of these clients.

# Create a boolean mask DataFrame for outliers in selected columns
outlier_mask = pd.DataFrame(False, index=df_2.index, columns=cols_with_outliers)

for col in cols_with_outliers:
    q1 = df_2[col].quantile(0.25)
    q3 = df_2[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 3 * iqr
    upper = q3 + 3 * iqr
    outlier_mask[col] = (df_2[col] < lower) | (df_2[col] > upper)

# Add Client_ID for reference
outlier_mask["Client_ID"] = df_2["Client_ID"]

# Count how many outlier columns each client has
outlier_counts = outlier_mask[cols_with_outliers].sum(axis=1)

# Filter for clients with more than 1 outlier column
multi_outlier_clients = outlier_mask.loc[outlier_counts > 1, "Client_ID"].unique()

print(f"Clients with outliers in more than one column: {multi_outlier_clients}")

# Optional: DataFrame showing their outlier flags per column
multi_outlier_df = outlier_mask[outlier_mask["Client_ID"].isin(multi_outlier_clients)]
print(multi_outlier_df)
```

### Finding: Error in Original Spend Variables

In the original dataset, the spend variables were inconsistent:

- **Spend_Total** was defined as the sum of 5 categories (Wines, Fruits, Meat, Fish, Sweets) but **did not include Gold**.  
- **Spend_RegularProducts** was defined as `Spend_Total - Spend_Gold`.  
- This created a logical error, because Gold was **never part of Spend_Total** in the first place.  

**Implication:**  
- The naming was misleading and inconsistent.  
- What was called `Spend_Total` was actually a subtotal (5 categories only).  
- `Spend_RegularProducts` was incorrectly derived from that subtotal.  

We corrected this by redefining:  
- `Spend_Total` = Wines + Fruits + Meat + Fish + Sweets + Gold  
- `Spend_RegularProducts` = Wines + Fruits + Meat + Fish + Sweets  

```python
# Build comparison table (copy only, df_2 unchanged)
check_campaigns = df_2[[
    "Total_Campaigns_Accepted",
    "LastCampaign_Response"
]].copy()

# Calculate accepted campaigns WITHOUT last campaign
check_campaigns["Total_Campaigns_Calc"] = (
    df_2["Campaign1_Response"] + df_2["Campaign2_Response"] +
    df_2["Campaign3_Response"] + df_2["Campaign4_Response"] +
    df_2["Campaign5_Response"]
)

# Add a difference column
check_campaigns["Diff"] = (
    check_campaigns["Total_Campaigns_Accepted"] - check_campaigns["Total_Campaigns_Calc"]
)

# Show column totals
totals = check_campaigns.sum().to_frame(name="Sum").T
print(totals)

# Show first 10 rows
print(check_campaigns.head(10))
```

```python
from scipy.stats import pointbiserialr
import pandas as pd

# Define binary and numeric variables
binary_cols = [
    "Campaign1_Response", "Campaign2_Response", "Campaign3_Response", 
    "Campaign4_Response", "Campaign5_Response", "LastCampaign_Response", 
    "Had_Complaint", "Marital_Divorced", "Marital_Married", 
    "Marital_Single", "Marital_Together", "Marital_Widow", 
    "Edu_2ndCycle", "Edu_Basic", "Edu_Graduation", "Edu_Master", "Edu_PhD"
]

num_cols = [
    "Household_Income", "Kids_Count", "Teens_Count", "Days_Since_Last_Purchase", 
    "Spend_Wines", "Spend_Fruits", "Spend_Meat", "Spend_Fish", 
    "Spend_Sweets", "Spend_Gold", "Purchases_Deals", "Purchases_Web", 
    "Purchases_Catalog", "Purchases_Store", "Web_Visits_Month", 
    "Customer_Age", "Days_Since_Became_Customer", "Spend_Total", 
    "Spend_RegularProducts", "Total_Campaigns_Accepted"
]

results = []
for b in binary_cols:
    for n in num_cols:
        try:
            corr, p = pointbiserialr(df_2[b], df_2[n])
            results.append((b, n, corr, p))
        except Exception as e:
            results.append((b, n, None, None))

# Build results DataFrame
corr_table = pd.DataFrame(results, columns=["Binary", "Numeric", "Correlation", "p_value"])

# Absolute correlation strength for easy sorting
corr_table["AbsCorr"] = corr_table["Correlation"].abs()

# Sort by strongest relationships
corr_table_sorted = corr_table.sort_values("AbsCorr", ascending=False)

# Show top 20 strongest binary-numeric relationships
print(corr_table_sorted.head(50))
```
